```python
# Calculate the IQR for 'windspeed'
Q1 = df['windspeed'].quantile(0.25)
Q3 = df['windspeed'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['windspeed'] < lower_bound) | (df['windspeed'] > upper_bound)]

# Print outliers
print('Outliers in Windspeed:')
print(outliers)
```

```python
from scipy import stats

# Hypothesis Testing
# 1. Working Day has an effect on number of electric cycles rented
# Null Hypothesis (H0): Working Day has no effect on number of electric cycles rented
# Alternate Hypothesis (H1): Working Day has an effect on number of electric cycles rented

# Split data based on working day
working_day = df[df['workingday'] == 1]['count']
non_working_day = df[df['workingday'] == 0]['count']

# Perform independent t-test
t_stat, p_val = stats.ttest_ind(working_day, non_working_day)

# Print results
print("T-Test Result:")
print(f"t-statistic: {t_stat:.4f}")
print(f"p-value: {p_val:.4f}")
```

* **Questions :** check if Working Day has an effect on the number of electric cycles rented

* **Answer :** based on data we have and the t-test performed above, there is not enough evidence to conclude that the working day has a signi cant effect on the number of electric cycles rented.

* **Insights  :** The number of electric cycles rented does not signi cantly differ between working days and non-working days. This could mean that the demand for cycles is relatively stable throughout the week.

* **Recommendations :** When planning the supply of cycles, consider that the demand does not signi cantly change between working days and non-working days. Therefore, a consistent supply of cycles could be maintained throughout the week. Further investigate other factors that could affect cycle demand, such as weather conditions, holidays, local events, and so on. Understanding how these variables interact with working days could provide valuable insights for improving service and increasing cycle usage.

```python
from scipy import stats

# 2. No. of cycles rented similar or different in different seasons
# Null Hypothesis (H0): No. of cycles rented is similar in different seasons
# Alternate Hypothesis (H1): No. of cycles rented is different in different seasons


# Separate the 'count' values by season
spring = df[df['season'] == 1]['count']
summer = df[df['season'] == 2]['count']
fall = df[df['season'] == 3]['count']
winter = df[df['season'] == 4]['count']

# Perform one-way ANOVA
f_stat, p_val = stats.f_oneway(spring, summer, fall, winter)

# Print results
print("ANOVA Result:")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")
```

* **Questions :** Check if No. of cycles rented is similar or different in different
  1. weather
  2. season

* **Answer :** based on the data we have and the ANOVA test performed above, there is strong evidence to conclude that the number of electric cycles rented is different in different seasons.

* **Insights  :** The number of cycles rented varies signi cantly across different seasons. This could mean that certain seasons are more favorable for cycle rentals than others. For example, warmer seasons might see a higher demand for cycles compared to colder seasons.

* **Recommendations :** When planning the supply of cycles, consider the seasonality of demand. For instance, more cycles could be made available in seasons with higher demand. Further investigate the relationship between season and cycle demand. Understanding how these variables interact could provide valuable insights for improving service and increasing cycle usage.
```python
import pandas as pd
from scipy import stats

# Hypothesis Testing
# Null Hypothesis (H0): Weather is independent of season
# Alternate Hypothesis (H1): Weather is dependent on season

# Create contingency table
contingency_table = pd.crosstab(df['weather'], df['season'])

# Perform Chi-Square Test
chi2, p_val, dof, expected = stats.chi2_contingency(contingency_table)

# Print the results
print("Chi-Square Test Result:")
print(f"Chi2 Statistic : {chi2:.4f}")
print(f"P-value        : {p_val:.4f}")
print(f"Degrees of Freedom: {dof}")
```

* **Questions :** check if Weather is dependent on the season

* **Answer :**  based on the data we have and the Chi-Square test performed above, there is strong evidence to conclude that the weather is dependent on the season.

* **Insights  :** The weather conditions are not distributed equally across the seasons. This could mean that certain weather conditions are more likely to occur in certain seasons. For example, clear or partly cloudy weather might be more common in summer, while heavy rain or snow might be more common in winter.
